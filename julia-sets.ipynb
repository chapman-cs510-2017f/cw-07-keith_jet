```python
import numpy as np
from cplane_np_hw import JuliaPlane as JP
import matplotlib.pyplot as plt

# c values
arrC = [0, -1, 0.3, -1.037 + 0.17*1j,-0.624 + 0.435*1j,-0.8 + 0.2*1j,-0.835 - 0.2321*1j]

# create sub plots
fig, axes = plt.subplots(2, 4, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})


fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, c in zip(axes.flat, arrC):
    j=JP(c)
    ax.imshow(j.plane, interpolation='bicubic', cmap='viridis')
    ax.set_title('c='+str(c))

plt.show()


```
